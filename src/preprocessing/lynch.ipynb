```python
import pandas as pd
import numpy as np
```

```python
cgm = pd.read_csv('../../data/processed/cgm/lynch2022.csv')
```

```python
cgm.head()
```

```python
peeps = cgm[['id', 'age', 'sex', 'type']].drop_duplicates()
```

```python
peeps
```

```python
lady_peeps = peeps[(peeps['sex']=='F')&(peeps['type']==1)]
```

```python
lady_peeps.sort_values('id')
```

```python
lady_peeps_cgm_set = lady_peeps['id'].unique()
```

## Menstrual cycle

```python
mi = pd.read_csv('../../data/raw/Lynch2022/DataTables/IOBP2PregMenstrInfo.txt', sep='|')
```

```python
mi_periods = mi[
    (pd.notna(mi['LastMenstrDt']) & (mi['LastMenstrDt'] != '')) |
    (pd.notna(mi['LastMenstrDtApprox']) & (mi['LastMenstrDtApprox'] != ''))
]
```

```python
mi_periods.sort_values('PtID')
```

```python
ids_with_periods = mi_periods['PtID'].unique()
```

```python
ids_with_periods
```

```python
lady_peeps_cgm_set
```

## Participants with CGM and periods

```python
lady_peeps_periods = set(lady_peeps_cgm_set).intersection(set(ids_with_periods))
```

```python
len(lady_peeps_periods)
```

## CGM Subset

```python
reading_counts = cgm[cgm['id'].isin(lady_peeps_periods)].groupby('id').size()
```
